```python
import dexplot as dxp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("survey.csv")
data.head()
```

#Gender Diversity in the Survey

```python
dxp.count("Gender",data=data,orientation='h')
```

Favourite Online Shopping Website

```python
dxp.count("Which online shopping website do you use the most?",data=data,orientation='h')
```

Recomended Product by a Friend

1 = 25% chance of buying
2 = 50% chance of buying
3 = 75% chance of buying
4 = 100% chance of buying

```python
dxp.hist("friend_recomend",data=data)
```

Social Media for Product Ads

```python
dxp.hist("Social_Media ",data=data)
```

Prefered Payment Method

```python
dxp.count('Which payment option do you prefer the most while making an online purchase?',data=data,orientation='v')
```

Pre-Online Payment

```python
dxp.count(" Do you prefer pre-online payment if there is an extra discount on it?",data=data)
```

Delivery Charges Vs Price of Item

```python
dxp.kde(x = "Price of the item (comparatively lower than the price that is offered in offline markets)",y = "Delivery Charges",data=data)
```
